# Pandas = pd

```python
import numpy as np
import pandas as pd
```

```python
# Creamos y poblamos un array NumPy de 5x2.
my_data = np.array([[0, 3], [10, 7], [20, 9], [30, 14], [40, 15]])

# Creamos una lista de python que contenga los nombres de las columnas
my_columns_names = ["temperature", "activity"]

# Creamos el DataFrame
my_dataframe = pd.DataFrame(data=my_data, columns=my_columns_names)

# Imprimimos el DataFrame entero
print(my_dataframe)
```

# Añadimos una nueva columna

```python
# Creamos una nueva columna llamada "adjusted"
my_dataframe["adjusted"] = my_dataframe["activity"] + 2
my_dataframe["whathever"] = my_dataframe["adjusted"] / 2

print(my_dataframe)
```

# Especificando un subset de datos

```python
print("Rows #0, #1, and #2:")
print(my_dataframe.head(3), "\n")

print("Row #2:")
print(my_dataframe.iloc[[2]], "\n")

print("Rows #1, #2, and #3:")
print(my_dataframe[1:4], "\n")

print("Column \"temperature\":")
print(my_dataframe["temperature"])
```

# Copiando un DataFrame

  • **Referenciar:** Si asignas un data frame a una variable, cualquier cambio en el DataFrame o en la nueva variable afectará al otro.

  • **Copia:** Si llamamos al método "pd.DataFrame.copy", creamos una copia independiente real. Los cambios en el DataFrame original o en la caopia no afectaran el uno en el otro.

```python
# Creamos una referencia asignando my_dataframe a una nueva variable
print("Experiment with a reference:")
reference_to_df = df

# Imprimir el valor inicial de una celda particular
print("   Starting value of df: %d" % df["Jason"][1])
print("   Starting value of reference_to_df: %d\n" % reference_to_df["Jason"][1])

# Modificamos una celda en "df"
df.at[1, "Jason"] = df["Jason"][1] + 5
print("   Updated df: %d\n\n" % reference_to_df["Jason"][1])

# Crear una copia real de my_dataframe
print("Experiment with a true copy:")
copy_of_my_dataframe = my_dataframe.copy()

# Imprimir el valor inincial de una celda particular
print("   Starting value of my_dataframe: %d\n" % my_dataframe["activity"][1])
print("   Starting value of copy_of_my_dataframe: %d\n" % copy_of_my_dataframe["activity"][1])

# Modificamos una celda en df
my_dataframe.at[1, "activity"] = my_dataframe["activity"][1] + 3
print("   Updated my_dataframe: %d" % my_dataframe["activity"][1])
print("   copy_of_my_dataframe does not get updated: %d" % copy_of_my_dataframe["activity"][1])
```
